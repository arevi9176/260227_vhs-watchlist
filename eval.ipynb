```python
# Beispiel für das lesen von key, value Paaren aus einem Dictionary
d = {"key1": "value1", "key2": "value2"}
for key, value in d.items():
    print(key, value)
```

```python
# Erstellen eines neuen Portfolio-Managers
pm = PortfolioManager("Portfolio_2")

# Iteration über die Daten und Hinzufügen von Portfolio-Elementen zum Portfolio-Manager
for name, details in data.items():
    print(name, details["isin"], details["stueckzahl"], details["kurs_beobachtung"])
    pm.add(PortfolioElement(details["isin"], details["stueckzahl"], details["kurs_beobachtung"]))
```
