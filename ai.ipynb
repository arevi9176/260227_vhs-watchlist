```python
import json

with open("secrets.json") as f:
    secrets = json.load(f)
    
GOOGLE_API_KEY = secrets.get("GOOGLE_API_KEY")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    api_key=GOOGLE_API_KEY,
)
```

```python
from langchain_core.messages import HumanMessage
human_msg = HumanMessage(content="Wie ist das aktuelle Wetter in Aachen?")

messages = [human_msg]
response = llm.invoke(messages)

print(response.content)
```

```python
from langchain_core.messages import SystemMessage

SHARE = "iShares Core MSCI World UCITS ETF USD Acc (ISIN IE00B4L5Y983)"

system_msg = SystemMessage("Du bist ein Experte für Aktien, ETF's und Anlagestrategien.")
human_msg = HumanMessage(f"Bewerte die folgende Aktie/ETF in Bezug auf kurzfristiges, mittelfristiges und langfristiges Potential: {SHARE}")

messages = [system_msg, human_msg]
response = llm.invoke(messages)
```

```python
from rich.console import Console
from rich.markdown import Markdown

console = Console()
md = Markdown(response.content)
console.print(md)
```
